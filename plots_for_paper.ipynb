```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfd = pd.read_csv('./paper_plots/He3_50K/ud.csv')
dfod = pd.read_csv('./paper_plots/He3_50K/uod.csv')
```

```python
plt.figure()
plt.plot(dfd['q'], dfd['u00'], linewidth=4, color='black', label=r'$u_{00}$')
plt.plot(dfod['q'], dfod['u10'], linewidth=4, color='blue', linestyle='--', label=r'$u_{10}$')
plt.plot(dfod['q'], dfod['u11'], linewidth=4, color='red', linestyle=':', label=r'$u_{11}$')
plt.xlim(0.8,2);
plt.ylim(-1.5,1);
plt.xticks(fontsize=14);
plt.yticks(fontsize=14);
plt.legend(loc='lower right', handlelength=3.6, fontsize=16);
plt.ylabel(r'$u_{ij}$', fontsize=20);
plt.xlabel(r'$q(\sigma)$', fontsize=18);
plt.title('First order functions', fontsize=16)
filename='./paper_plots/He3_50K/He3_T50K_1stO.pdf'
plt.tight_layout()
plt.savefig(fname=filename,  bbox_inches="tight")
```

```python
plt.figure()
plt.plot(dfod['q'], dfod['u20'], linewidth=4, color='green', linestyle='-', label=r'$u_{20}$')
plt.plot(dfod['q'], dfod['u21'], linewidth=4, color='orange', linestyle='--', label=r'$u_{21}$')
plt.plot(dfod['q'], dfod['u22'], linewidth=4, color='magenta', linestyle=':', label=r'$u_{22}$')
plt.xlim(0.8,2);
plt.ylim(-1.5,1);
plt.xticks(fontsize=14);
plt.yticks(fontsize=14);
plt.legend(loc='lower right', handlelength=3.6, fontsize=16);
plt.ylabel(r'$u_{ij}$', fontsize=20);
plt.xlabel(r'$q(\sigma)$', fontsize=18);
plt.title('Second order functions', fontsize=16);
filename='./paper_plots/He3_50K/He3_T50K_2ndO.pdf'
plt.tight_layout()
plt.savefig(fname=filename,  bbox_inches="tight")
```
