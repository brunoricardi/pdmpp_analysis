```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./wave_cut_r5_lambda1212.csv")
```

```python
plt.figure(figsize=(10,8))
plt.title('Highest-momentum contributing wave for different cutoffs', fontsize=16)
plt.plot(df['tau'], df['l_pwc-3'], lw=4, label='$10^{-3}$')
plt.plot(df['tau'], df['l_pwc-5'], lw=4, label='$10^{-5}$')
plt.plot(df['tau'], df['l_pwc-7'], lw=4, label='$10^{-7}$')
plt.xlabel(r'$\tau$', fontsize=28)
plt.ylabel(r'$\ell_{cut}$', fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=16)
plt.xscale('log')
plt.yscale('log')
```
