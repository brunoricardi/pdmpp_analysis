```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename='./ud2.csv'
uij1 = pd.read_csv(filename)
filename='./ud3sq.csv'
uij2 = pd.read_csv(filename)
filename='./ud2sq.csv'
uij3 = pd.read_csv(filename)
filename='./ud1sq.csv'
uij4 = pd.read_csv(filename)
filename='./ud1.csv'
uij5 = pd.read_csv(filename)
```

```python
plt.figure(figsize=(8,6))
plt.plot(uij1['q']*2.56,uij1['up'],label="4sq", lw=3)
plt.plot(uij2['q']*2.56,uij2['up'],label="3sq", lw=3)
plt.plot(uij3['q']*2.56,uij3['up'],label="2sq", lw=3, linestyle='dashed')
plt.plot(uij4['q']*2.56,uij4['up'],label="1sq", lw=3, linestyle='dashed')
plt.plot(uij5['q']*2.56,uij5['up'],label="0sq", lw=3, linestyle='dashed')
#plt.xlim(0.78,1.57)
plt.xlim(2,4)
plt.ylim(-1.5,1.0)
plt.xlabel(r'$r (\AA)$', fontsize=20)
plt.ylabel(r'$u_{00}(r)$', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=14)
```
